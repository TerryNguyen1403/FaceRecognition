```python
alg = "C:/Users/nguye/Downloads/FaceRecognition-master/FaceRecognition-master/models/haarcascade_frontalface_default.xml"
face_detection = cv2.CascadeClassifier(alg)
```

```python
# Face Recognition 
def face_detect(img):
    face_img = img.copy()
    face_rects = haar_cascade.detectMultiScale(face_img)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(0,255,0),3)
    return face_img
```

```python
result = face_detect(img1)
```

```python
# Face Recognition 
def face_detection(img):
    face_img = img.copy()
    face_rects = haar_cascade.detectMultiScale(face_img, scaleFactor=1.2,minNeighbors=5)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(0,255,0),3)
    return face_img
```
